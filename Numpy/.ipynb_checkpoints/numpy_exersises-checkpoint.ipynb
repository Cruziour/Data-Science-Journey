## Valid Sudoku

```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
b = np.sum(s, axis=1)
b == 45
```

```python
for i in b:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("For rows it is valid")
```

```python
c = np.sum(s, axis=0)
for i in c:
    if i != 45:
        print("Sudoku is not valid")
else:
    print("For column it is valid")
```

```python
p = s[0:3, 0:3]
p
```

```python
p1 = s[0:3, 3:6]
p1
```

```python
p2 = s[0:3, 6:9]
p2
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n= s[i:i+3, j:j+3]
        print(n.sum())
```

## General Questions

```python
# Columns: Age, Math marks, Science Marks
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

#### Get the shape of the matrix

```python
data.shape
```

#### find the average age of students

```python
np.mean(data[:,0])
```

```python
data[:,0]
```

#### Extract Math marks of all students

```python
math_marks = data[:, 1:2]
sum_math = math_marks
math_marks
```

```python
np.sum(sum_math)
```

#### Find the highest science marks

```python
science_marks = data[:, 2]
science_marks
```

```python
np.max(science_marks)
```

#### Get details of the students who scored more than 90 in Math

```python
data[data[:, 1] >= 90]
```

#### Increase Math marks of all students by 4

```python
data[:,1] += 4
data
```

#### Find how many students are younger than 19

```python
younger_student = data[data[:, 0] < 19]
younger_student
```

```python
len(younger_student)
```

#### Calculate the average marks in each subject column-wise mean

```python
data[:, 1:]
```

```python
np.mean(data[:,1:], axis=0)
```

#### Get data of students who scored at least 80 in both subject

```python
data[:, 1:]
```

```python
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```
