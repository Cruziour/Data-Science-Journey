NUMPY INDEXING AND SLICING OF VECTORS

```python
arr = np.arange(11,21)
arr
```

```python
arr[6]
```

```python
arr[1:8]
```

```python
arr
```

```python
arr[1:8:2]
```

NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[0,1]
```

```python
arr[0:2,1:3]
```

```python
arr[3:6,3:5]
```

BOOLEAN INDEXING

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
