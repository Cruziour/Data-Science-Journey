#### Broadcasting

```python
l = [10,20,30,40]
arr = np.array(l)
```

```python
for i in l:
    print(i + 10)
```

```python
arr + 10
```

```python
arr2 = np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2 + 10
```

```python
arr2 * 5
```

##### Stacking Array

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

##### Splitting array

```python
c = np.arange(16).reshape(4,4)
c
```

```python
g = np.hsplit(c,4)
```

```python
p = np.vsplit(c, 4)
```

```python
for i in p:
    print(i)
```

```python
for i in g:
    print(i)
```
