# Pandas Series 
###### A series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.
### Creating Series
###### There are multiple ways to create a pandas Series.

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_lists = [10,20,30]
my_arr = np.array([10,20,30])
my_dict={'a':10, 'b':20, 'c':30}
```

```python
pd.Series(my_lists)
```

```python
pd.Series(my_lists, index=labels)
```

```python
pd.Series(my_arr, index=labels)
```

```python
pd.Series(my_dict)
```
